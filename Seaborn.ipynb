```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib as mpl
import seaborn as sns
import os 
%matplotlib inline
```

```python
Car = pd.read_csv('Vechile_Data.csv')
Car.head()
```

```python
brandVSfuel=Car.groupby(["make","fuel-type"]).size().unstack(fill_value=0)
brandVSfuel
```

```python
brandVSfuel.plot(kind='bar',stacked=True)
plt.show()
```

```python
brandVSdoors=Car.groupby(["make","fuel-type"]).size().unstack(fill_value=0)
brandVSdoors
```

```python
brandVSdoors.plot(kind='bar',stacked=False)
plt.show()
```

```python
Car.columns
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=Car["wheel-base"],linewidth=2,label="engine-type")
sns.lineplot(data=Car["engine-size"],linewidth=2,label="engine-size")
plt.title("wheel-base VS engine-size")
plt.ylabel("wheel-base")
plt.xlabel("engine-size")
plt.show()
```

```python
plt.figure(figsize=(22,10))
sns.lineplot(data=Car["body-style"],linewidth=2.5,label="horsepower")
sns.lineplot(data=Car["compression-ratio"],linewidth=2.5,label="fuel-system")
plt.title("body-style VS compression-ratio")
plt.ylabel("body-style")
plt.xlabel("horsepower")
plt.show()
```

```python
plt.figure(figsize=(22,10))
sns.set(rc={"axes.facecolor":"#FFFFFF","axes.grid":False,"xtick.labelsize":12,"ytick.labelsize":12})
plt.title("wheel-base VS engine-size")
plt.ylabel("wheel-base")
plt.xlabel("engine-size")
plt.xticks(rotation=60)
sns.lineplot(x=Car["engine-size"], y=Car["wheel-base"].index.values, color='r')
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.set(rc={"axes.facecolor":"#FFFFFF","axes.grid":True,"xtick.labelsize":12,"ytick.labelsize":10})
plt.title("body-style VS compression-ratio")
plt.ylabel("horsepower")
plt.xlabel("body-style")
plt.xticks(rotation=90)
sns.lineplot(x=Car["horsepower"], y=Car["body-style"].index.values, color='b')
plt.show()
```
