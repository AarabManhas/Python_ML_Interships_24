```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib as mpl
import seaborn as sns
import os 
%matplotlib inline
```

```python
Dt = pd.read_csv('Titanic.csv')
Dt
```

```python
survivedvsalive=Dt.groupby(["survived","alive"]).size().unstack(fill_value=0)
survivedvsalive
```

```python
 survivedvsalive.plot(kind='bar',stacked=False)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='survived', data=Dt)
plt.title('Survival Counts')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='class', data=Dt)
plt.title('Fare by Class')
plt.xlabel('Class')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='embark_town', data=Dt)
plt.title('Count of Embarked Towns')
plt.xlabel('Embark Town')
plt.ylabel('Count')
plt.show()
```
