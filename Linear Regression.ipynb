```python
 import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
np.random.seed(0)
Eng_s = np.random.normal(4.0,0.5,100)
Price=1000+Eng_s*1000+np.random.normal(0,2000,100)
```

```python
Eng_s
```

```python
Price
```

```python
Dt=pd.DataFrame({"Engine Size":Eng_s,"Price":Price})
Dt
```

```python
X=Dt[["Engine Size"]]
X
```

```python
Y=Dt["Price"]
Y
```

```python
model=LR()
model.fit(X,Y)
```

```python
y=model.predict(X)
```

```python
y
```

```python
MSE=mean_squared_error(Y,y)
MAE=mean_absolute_error(Y,y)
R2=r2_score(Y,y)
```

```python
MSE
```

```python
print(f"Mean squared Error={MSE}\nMeanAbsolute Error={MAE}\nR2 Score={R2}")
```

```python
MAE
```

```python
R2
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X,Y, color='green',label="Actual Data")
plt.plot(X,y, color='orange', label="predicted Data")

plt.xlabel("Engine size")
plt.ylabel("Price")
plt.title("Linear Regression")
plt.legend()
plt.show()      
```

```python
np.random.seed(0)
Eng_s1 = np.random.normal(3.0,0.5,100)
Wheel_b=1000+Eng_s1*1000+np.random.normal(0,0.5,100)
amt=1000+Eng_s1*1000+np.random.normal(0,200,100)
```

```python
print (Eng_s1)
print(Wheel_b)
print(amt)
```

```python
Dt2=pd.DataFrame({"Engine Size":Eng_s1,"Wheel Base":Wheel_b,"Amount":amt})
Dt2
```

```python
X1=Dt2[["Engine Size","Wheel Base"]]
Y1=Dt2["Amount"]
```

```python
MLRModel = LR()
MLRModel.fit(X1,Y1)
```

```python
y1 = MLRModel.predict(X1)
MSE1 = mean_squared_error(Y1,y1)
MAE1=mean_absolute_error(Y1,y1)
R21=r2_score(Y1,y1)
print(f"mean squared error = {MSE1} \n Mean absolute error = {MAE1} \n R2 error = {R21}")
```

```python
x_surf, y_surf = np.meshgrid(
    np.linspace(Dt2['Engine Size'].min(), Dt2['Engine Size'].max(), 100),
    np.linspace(Dt2['Wheel Base'].min(), Dt2['Wheel Base'].max(), 100)
)
```

```python
x_surf
```

```python
y_surf
```

```python
xy_surf=np.c_[x_surf.ravel(),y_surf.ravel()]
xy_surf
```

```python
z_surf=MLRModel.predict(xy_surf).reshape(x_surf.shape)
```

```python
z_surf
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(Dt2["Engine Size"], Dt2["Wheel Base"], Dt2["Amount"], color='blue', label="Actual data")


ax.plot_surface(x_surf, y_surf, z_surf, color="red", alpha=0.5, label="Predicted value")

ax.set_xlabel("Engine Size")
ax.set_ylabel("Wheel Base")
ax.set_zlabel("Amount")
ax.set_title("Engine size and Wheel base vs Price")
ax.legend()
plt.show()
```
